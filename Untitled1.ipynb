```python
import gender_guesser.detector as gender
from nameparser import HumanName as hn
# Initialize the gender detector
d = gender.Detector(case_sensitive=True)

# Function to predict gender based on first name
def predict_gender(name):
    name = hn(name)
    first_name = name.first # Get the first name
    return d.get_gender(first_name)

# Apply the function to the 'name' column
data['gender'] = data['name'].apply(predict_gender)


def classify(data):
    female_data = data[data['gender'].isin(['female', 'mostly_female', 'unknown'])]
    male_data = data[data['gender'].isin(['male', 'mostly_male'])]

    # Dropping the 'gender' column in both dataframes
    male_data = male_data.drop('gender', axis=1)
    female_data = female_data.drop('gender', axis=1)

    # Printing the counts
    print(f"females :\t{female_data.shape[0]} \nmales :\t{male_data.shape[0]}")
    #female_data.to_csv(f"C://Users/CPS/Documents/cleaned_leads_female/leads_{pd.Timestamp.now().date()}.csv")
    male_data.to_csv(f"C://Users/CPS/Documents/cleaned_leads_male/leads_{pd.Timestamp.now().date()}.csv")
    return female_data, male_data 

# Assuming 'data' is your pandas DataFrame
# Call the function with your data
female_data, male_data = classify(data)
```
