```python
# checking for number of unique values in each column
# those columns having only 1 unique value should be removed 
def uni_check_cleaner():  
    for i in data.columns:
        if len(data[i].unique()) == 1:
            data.drop(columns = i, axis = 1, inplace = True)
        else:
            print(f"{i}\t {len(data[i].unique())}")
    print(data.shape[1])
        
```
