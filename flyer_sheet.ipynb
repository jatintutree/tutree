```python
import pandas as pd
import qrcode
import os

df = pd.read_csv("C://Users/CPS/Downloads/flyer_sheet - texas_district_flyer.csv")

# Directory to save the QR codes
output_folder = "C://Users/CPS/Documents/qr code flyers/qr_code_texas/"
os.makedirs(output_folder, exist_ok=True)

# Generate and save QR codes using the school name
for index, row in df.iterrows():
    school_name = row["district"].replace(" ", "_").replace("/", "_").replace("\\", "_")  # Replace spaces with underscores for file naming
    url = row["ppc_link"]
    
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    qr.add_data(url)
    qr.make(fit=True)
    
    img = qr.make_image(fill="black", back_color="white")
    file_path = os.path.join(output_folder, f"{school_name}_QR.png")
    img.save(file_path)

output_folder
```

```python
import pandas as pd
from PyPDF2 import PdfReader, PdfWriter

pdfs = pd.read_csv("C://Users/CPS/Desktop/flyer_path.csv")

for index, row in pdfs.iterrows():
    eng_file = row['eng']
    spn_file = row['spn']
    
    eng_reader = PdfReader(eng_file)
    spn_reader = PdfReader(spn_file)
    
    merger = PdfWriter()
    
    for page in eng_reader.pages:
        merger.add_page(page)
    
    for page in spn_reader.pages:
        merger.add_page(page)
    
    output_path = f"C://Users/CPS/Documents/result/merged_{index+1}.pdf"
    
    with open(output_path, "wb") as output_file:
        merger.write(output_file)
    print(f"Merged PDF {index+1} saved as {output_path}")

print("All PDFs have been successfully merged.")
```
