```python
import requests
import pandas as pd
import json

# Define the base URL for the API
base_url = "http://104.225.218.197:31102/linkedin-knn/"

df = pd.read_csv("C://Users/CPS/Downloads/Untitled discover search.csv")
random_rec = df.sample(n=15)
random_rec.head(15)

# Initialize a list to store the API responses
api_responses = []

# Loop through each LinkedIn username and make the API call
for username in random_rec['linkedin_username']:
    api_url = f"{base_url}{username}"
    
    try:
        response = requests.get(api_url)
        response.raise_for_status()
        
        # Parse the JSON response and store it in the list
        api_responses.append(json.dumps(response.json()))
    
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve data for {username}: {e}")
        api_responses.append(None)


df = pd.DataFrame(api_responses)
df.head()
```

```python
import requests
import pandas as pd

# Define the base URL for the API
base_url = "http://104.225.218.197:31102/linkedin-knn/"

# Load your CSV file
df = pd.read_csv("C://Users/CPS/Downloads/Untitled discover search.csv")

# Select a random sample of 15 records
random_rec = df.sample(n=15)

# Initialize a list to store the raw API responses
api_responses = []

# Loop through each LinkedIn username and make the API call
for username in random_rec['linkedin_username']:
    api_url = f"{base_url}{username}"
    
    try:
        response = requests.get(api_url)
        response.raise_for_status()
        
        # Store the raw text response in the list
        api_responses.append(response.text)
    
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve data for {username}: {e}")
        api_responses.append(None)



df = pd.DataFrame(api_responses, columns=['api_responses'])
df.head()
```

```python
import matplotlib.pyplot as plt

# Filter the relevant columns: 'score' and 'industry'
score_data = data[['score', 'industry']]

# Group by industry and calculate the average score for each industry
average_scores = score_data.groupby('industry').mean().sort_values(by='score', ascending=False)

# Plotting the average score per industry
plt.figure(figsize=(10, 6))
plt.barh(average_scores.index, average_scores['score'], color='skyblue')
plt.xlabel('Average Score')
plt.ylabel('Industry')
plt.title('Average Score by Industry')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
# Creating the three graphs: Box Plot, Scatter Plot, and Histogram with Industry column

import seaborn as sns

# Box Plot with Industry
plt.figure(figsize=(10, 6))
sns.boxplot(x='industry', y='score', data=data)
plt.xticks(rotation=90)
plt.title('Box Plot of Scores by Industry')
plt.xlabel('Industry')
plt.ylabel('Score')
plt.show()
```

```python


# Scatter Plot with Industry
plt.figure(figsize=(10, 6))
sns.scatterplot(x=range(len(data)), y='score', hue='industry', data=data, palette='Set1', alpha=0.7)
plt.title('Scatter Plot of Scores by Industry')
plt.xlabel('Index')
plt.ylabel('Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()


```

```python
# Histogram with Industry
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='score', hue='industry', multiple='stack', palette='Set1', edgecolor='black', bins=20)
plt.title('Histogram of Scores by Industry')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```
